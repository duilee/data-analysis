```python
# crawl 멜론 노래 순위
url = "http://www.melon.com/chart/index.htm"
driver.get(url)

#beautiful soup 으로 읽고 parse
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
songs = soup.select('table > tbody > tr')
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    print(title, singer, sep = ' | ')
    
# with just selenium
"""
songs = driver.find_elements_by_css_selector('table > tbody > tr')
for song in songs:
    title = song.find_elements_by_css_selector('div.ellipsis.rank01 > span > a')[0].text
    singer = song.find_elements_by_css_selector('div.ellipsis.rank02 > a')[0].text
    print(title, singer, sep =' | ')

"""
#selenium + BS4 is faster, but you cant take actions on the browser
# just selenium is a little slower but taking actions is allowed
```

```python
## make excel of ranks from melon, bugs, genie

song_data = []
rank = 1
songs = soup.select('table > tbody > tr')
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    song_data.append(['Melon', rank, title, singer])
    rank += 1

columns = ['서비스', '순위', '타이틀', '가수']
pd_data = pd.DataFrame(song_data, columns = columns)
pd_data.to_excel('./files/melon.xlsx', index=False)

pd_data.head()
```

```python
#top 100 from bugs
song_data = []
rank = 1

url = 'https://music.bugs.co.kr/chart'
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

songs = soup.select('table.byChart > tbody > tr')
for song in songs:
    title = song.select('p.title > a')[0].text
    singer = song.select('p.artist > a')[0].text
    song_data.append(['Bugs', rank, title, singer])
    rank += 1
    
columns = ['서비스', '순위', '타이틀', '가수']
pd_data = pd.DataFrame(song_data, columns = columns)
pd_data.to_excel('./files/bugs.xlsx', index=False)
pd_data.head()
    
```

```python
# get top 50 from genie
song_data = []
rank = 1
url = "https://www.genie.co.kr/chart/top200"
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

songs = soup.select('table.list-wrap > tbody > tr')
for song in songs:
    title = song.select('a.title')[0].text.strip()
    singer = song.select('a.artist')[0].text
    song_data.append(['Genie', rank, title, singer])
    rank += 1
    
columns = ['서비스', '순위', '타이틀', '가수']
pd_data = pd.DataFrame(song_data, columns = columns)
pd_data.to_excel('./files/genie.xlsx', index=False)
pd_data.head()
```

```python
excel_names = ['./files/melon.xlsx', './files/bugs.xlsx', './files/genie.xlsx']
appended_data = pd.DataFrame()
for name in excel_names:
    pd_data = pd.read_excel(name)
    appended_data = appended_data.append(pd_data)

appended_data.to_excel('./files/total.xlsx')
appended_data.head()
```
